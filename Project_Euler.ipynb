## Largest palindrome product

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
def palindrome_solution_one_slowest():
    factor_list = []
    palindrome_list = []
    for x in range(10000,998002):  #100 x 100 = 10000 and 999 x 999 = 998002
        factor_list = []
        if str(x) == str(x)[::-1]:
            palindrome = x
            for factor in range(100,1000):
                if palindrome % factor == 0:
                    factor_list.append(factor)
                for x in factor_list:
                    for y in factor_list:
                        if x * y == palindrome:
                            if palindrome not in palindrome_list:
                                palindrome_list.append(palindrome)
                                break
        else:
            continue
    return sorted(palindrome_list, reverse=True)[0]

def palindrome_solution_two_faster():
    palindrome_list = []
    for x in range(100,1000):
        for y in range(100,1000):
            number = x * y
            if str(number) == str(number)[::-1]:
                palindrome_list.append(number)
    return sorted(palindrome_list, reverse=True)[0]

def palindrome_solution_one_line():
    return sorted([x*y for x in range(100,1000) for y in range(100,1000) if str(x*y) == str(x*y)[::-1]], 
       reverse=True)[0]
```

## Smallest Multiple

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.<br>
What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
def smallest_multiple():
    alist = []
    for multiple in range(100000000,500000000,2): # trial and error to find the correct range
        if bool(alist) == True: #stops after finding first (smallest), multiple
            break
        for x in range(11,21): # only need factors 11-20 because if true, factors 1-10 must also be true
            if multiple % x != 0:
                break
            if x == 20: #if algorithm reaches end of factor range without breaking, add answer to alist
                alist.append(multiple)
    return alist
```

```python
import functools
month_lengths_non_leap = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
month_lengths_leap = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# get all the first of the months
firsts_of_month_non_leap = [1]
for index, number in enumerate(month_lengths_non_leap):
    if index < 11:
        firsts_of_month_non_leap.append(functools.reduce(lambda x,y: x + y, month_lengths_non_leap[:index + 1]) + 1)
        
firsts_of_month_leap = [1]
for index, number in enumerate(month_lengths_non_leap):
    if index < 11:
        firsts_of_month_leap.append(functools.reduce(lambda x,y: x + y, month_lengths_leap[:index + 1]) + 1)

month_counter = 0
counter = 6
for year in range(1901,2001):
    if year % 4 == 0:
        for day in range(counter,366,7):
            if day in firsts_of_month_leap:
                month_counter += 1
        counter -= 2
        if counter == 0:
            counter = 7
    else:
        for day in range(counter,366,7):
            if day in firsts_of_month_non_leap:
                month_counter += 1
        counter -= 1
        if counter == 0:
            counter = 7
print(month_counter)
```
